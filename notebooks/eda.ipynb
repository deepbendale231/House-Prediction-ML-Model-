# EDA - California Housing Dataset

This notebook covers:
- Income category distribution
- Correlation heatmap
- Geographic scatter plot with median house value

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_path = Path('..') / 'data' / 'housing.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
df['income_cat'] = pd.cut(
    df['median_income'],
    bins=[0.0, 1.5, 3.0, 4.5, 6.0, float('inf')],
    labels=[1, 2, 3, 4, 5]
)
df['income_cat'].value_counts().sort_index()
```

```python
ax = df['income_cat'].value_counts().sort_index().plot(
    kind='bar',
    figsize=(8, 4),
    grid=True,
    color='steelblue'
)
ax.set_title('Income Category Distribution')
ax.set_xlabel('Income Category')
ax.set_ylabel('Count')
plt.tight_layout()
```

```python
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=['income_cat'])
corr = numeric_df.corr()

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(corr.values, cmap='coolwarm', vmin=-1, vmax=1)
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=90)
ax.set_yticklabels(corr.columns)
ax.set_title('Correlation Heatmap')
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
```

```python
ax = df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    alpha=0.35,
    s=df['population'] / 100,
    c='median_house_value',
    cmap='viridis',
    colorbar=True,
    figsize=(10, 7)
)
ax.set_title('Geographic Scatter (Color = Median House Value)')
plt.tight_layout()
```
