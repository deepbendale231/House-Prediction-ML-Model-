```python
import pandas as pd
import numpy as np
```

```python
# use a raw string or forward slashes to avoid backslash escape issues
df = pd.read_csv(r"C:\Users\deepb\Desktop\gurgaon city price prediction\data\housing.csv")
```

```python
df['income_cat'] = pd.cut(df['median_income'],
                          bins=[0., 1.5, 3.0, 4.5, 6., np.inf], 
                          labels=[1, 2, 3, 4, 5]) # create income category attribute
```

```python
import numpy as np  
```

```python
from sklearn.model_selection import StratifiedShuffleSplit 
```

```python
split=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) # create stratified split object
#n - number of re-shuffling & splitting iterations
# test_size - proportion of dataset to include in test split # random_state - for reproducibility
for train_index, text_index in split.split(df, df["income_cat"]):
    strat_train_set=df.loc[train_index] # create stratified training set
    strat_test_set=df.loc[text_index]

  
```

```python
# lets remove the income_cat column
for set in (strat_train_set, strat_test_set):
    set.drop("income_cat", axis=1, inplace=True) # inplace=True to modify the set directly
```

```python
df = strat_train_set.copy()
```

```python
df.plot(kind='scatter', x='longitude', y='latitude', grid=True , alpha=0.4)
        # looks like a map of california.
```
